#### We now want to compute the mean salary after removing the maxima for each Department. Create the appropriate pivot table.

Hint: Write a custom function.

```python
employee.pivot_table(index=['Department', 'Name'])
```

```python
# your answer here
def mean_wo_max(x):
    x = list(x)
    x.remove(max(x))
    return sum(x) / len(x)

means = employee.pivot_table(index=['Department'], values='Salary', aggfunc=mean_wo_max)
means
```
